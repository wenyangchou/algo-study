```python
import pandas as pd

df = pd.read_csv('iris.csv')
df.head()
```

```python
labels = df['species'].unique().tolist()
df['label'] = df['species'].apply(lambda x:labels.index(x)) 
x_columns = ['sepal_length','sepal_width','petal_length','petal_width']
y_columns = ['label']
```

```python
from catboost import CatBoostRegressor, Pool

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split


x = df[x_columns]
y = df[y_columns]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=6)
x_train.head()

# mdl = CatBoostRegressor(
#     objective = 'regression',
#     num_leaves = 31,
#     learning_rate = 0.05,
#     loss_function='RMSE',
#     verbose = -1)

mdl = CatBoostRegressor(num_leaves = 31,
                          learning_rate=0.05, 
                          loss_function='RMSE')
 
mdl.fit(x_train, y_train)
```

```python
mdl.save_model('model.cbm')
```

```python
test_pool = Pool(x_test) 

y_pred = mdl.predict(test_pool)
# eval
rmse_test = mean_squared_error(y_test, y_pred) ** 0.5
print(f'The RMSE of prediction is: {rmse_test}')
```
